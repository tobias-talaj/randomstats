```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/home/tobia/Repos/randomstats/ratebeer/beer_data.csv')
```

```python
df.loc[df.loc[:, 'city'] == 'Copenhagen', 'state'] = 'Copenhagen'
df.loc[df.loc[:, 'city'] == 'Copenhagen N', 'state'] = 'Copenhagen'
df.loc[df.loc[:, 'state'] == 'Copenhagen', 'city'] = 'Copenhagen'
```

```python
df_points_per_country = df.groupby(['country'])['points'].agg('sum').sort_values(ascending=False).head(25)
ax = df_points_per_country.plot.barh(x='country', y='points', figsize=[8, 10])
plt.gca().invert_yaxis()
plt.tick_params(labeltop=True)
plt.grid(axis='x')
plt.show()
```

```python
df_beers_per_country = df.loc[:, 'country'].value_counts().head(25)
ax = df_beers_per_country.plot.barh(x='country', y='points', figsize=[8, 10])
plt.gca().invert_yaxis()
plt.tick_params(labeltop=True)
plt.grid(axis='x')
plt.show()
```

```python
top50_cities_pts = df
top50_cities_pts.loc[:, 'city_with_state'] = top50_cities_pts.loc[:, 'city'] + ', ' + top50_cities_pts.loc[:, 'state']
top50_cities_pts = df.groupby(['city_with_state'])['points'].agg('sum').sort_values(ascending=False).head(50)
ax = top50_cities_pts.plot.barh(x='city_with_state', y='beers', figsize=[16, 24])
plt.gca().invert_yaxis()
plt.tick_params(labeltop=True)
plt.grid(axis='x')
plt.show()
```

```python
top50_regions_pts = df
top50_regions_pts.loc[:, 'region_with_country'] = top50_regions_pts.loc[:, 'state'] + ', ' + top50_regions_pts.loc[:, 'country']
top50_regions_pts = df.groupby(['region_with_country'])['points'].agg('sum').sort_values(ascending=False).head(50)
ax = top50_regions_pts.plot.barh(x='region_with_country', y='beers', figsize=[16, 24])
plt.gca().invert_yaxis()
plt.tick_params(labeltop=True)
plt.grid(axis='x')
plt.show()
```

```python
top50_breweries_pts = df
top50_breweries_pts.loc[:, 'brewery_with_country'] = top50_breweries_pts.loc[:, 'brewery'] + ', ' + top50_breweries_pts.loc[:, 'country']
top50_breweries_pts = df.groupby(['brewery_with_country'])['points'].agg('sum').sort_values(ascending=False).head(50)
ax = top50_breweries_pts.plot.barh(x='brewery_with_country', y='beers', figsize=[16, 24])
plt.gca().invert_yaxis()
plt.tick_params(labeltop=True)
plt.grid(axis='x')
# plt.savefig('top50_breweries.svg')
plt.show()
```
